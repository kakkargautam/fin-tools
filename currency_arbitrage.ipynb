```python
import pandas as pd 
import numpy as np
df = pd.read_csv("fx_rate_2.csv") 
print (df)
```

```python
import networkx as nx
G = nx.from_pandas_edgelist(df, source='Currency_A', target='Currency_B', edge_attr="Rate_To", create_using=nx.DiGraph())
```

```python
nx.draw(G,
        with_labels=True,
        node_color="skyblue"
       )
```
