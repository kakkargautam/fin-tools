Introduction 
Code to calculate moving averages.

```python
# Loading Apple weekly Data (from Yahoo )into dataframes
import pandas as pd 
df = pd.read_csv("AAPL-2.csv", parse_dates=['Date']) 
print (df[:5])
```

```python
# Calculating weekly returns and 3 weeek moving averages.
df ['weekly_ret'] = ( df['Adj Close'] / df['Adj Close'].shift(1) ) - 1
df ['rollRet_3week'] = df ['weekly_ret'].rolling(3).mean()
print (df[:5])
```

```python
from bokeh.plotting import figure
from bokeh.io import show, output_notebook
from bokeh.models import ColumnDataSource
p = figure(title = 'Apple',x_axis_type = 'datetime', plot_height=200, plot_width=800)
cds_obj = ColumnDataSource(df)
p.line(x= 'Date', y= 'weekly_ret', source = cds_obj)
p.line(x= 'Date', y= 'rollRet_3week', source = cds_obj, color= 'red')
output_notebook()
show(p)
```
